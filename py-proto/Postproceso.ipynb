```python
import pandas as pd

def fasta_to_series(f, file_type="fasta"):
    dict = []
    for i in SeqIO.parse(f, file_type):
        dict.append({ "id": i.id, "seq": i.seq, "name": i.name, "description": i.description})
    s = pd.Series(dict)
    return s

def all_files(file_type="fasta"):
    
    root = "/data/disk2/marta/SYNOLOGY"
    coll = ["07.Non-Collapsed", "06.Collapsed"]
    method = ["JB","WAVn1", "WAVn2"]

    path = []
    for r in os.listdir(root):
        for c in coll:
            for m in method:
                path.append(op.join(root,r,c,m))
    
    fasta_files = []
    for p in path:
        if not op.isdir(p):
            continue
        for l in os.listdir(p):
            if re.search(file_type, l):
                fasta_files.append(op.join(p,l))
    return fasta_files

def all_grouped_files(file_type="fasta", groups = ['RIVA', 'SUDHL10', '10M']):
    files = all_files(file_type)
    
    gr = []
    for g in groups:
        l = []
        for f in files:
            if re.search(g,f):
                l.append(f)
        gr.append(l)
    return gr

def group_tontology():
    
    gr = all_grouped_files(file_type="csv")
    
    
    all_files = []
    for g in gr:
        files = {}
        for f in g:
            if re.search('Non-Collapsed', f):
                if re.search('JB',f):
                    try:
                        files['non_JB'].append(f)
                    except:
                        files['non_JB'] = []
                        files['non_JB'].append(f)
                elif re.search('WAVn1', f):
                    try:
                        files['non_WAVn1'].append(f)
                    except:
                        files['non_WAVn1'] = []
                        files['non_WAVn1'].append(f)
                else: 
                    try:
                        files['non_WAVn2'].append(f)
                    except:
                        files['non_WAVn2'] = []
                        files['non_WAVn2'].append(f)
            else:
                if re.search('JB',f):
                    try:
                        files['JB'].append(f)
                    except:
                        files['JB'] = []
                        files['JB'].append(f)

                elif re.search('WAVn1', f):
                    try:
                        files['WAVn1'].append(f)
                    except:
                        files['WAVn1'] = []
                        files['WAVn1'].append(f)
                else: 
                    try:
                        files['WAVn2'].append(f)
                    except:
                        files['WAVn2'] = []
                        files['WAVn2'].append(f)
        
        all_files.append(files)
        
    return files

        
    

def xls_to_pandas(file):
    from openpyxl import load_workbook
    from itertools import islice
    xls = load_workbook(file)
    filename, _ = os.path.splitext(file)
    basename = op.basename(filename)
    
    data = xls[basename].values
    cols = next(data)[1:]
    data = list(data)
    idx = [r[0] for r in data]
    data = (islice(r, 1, None) for r in data)
    df = pd.DataFrame(data, index=idx, columns=cols)
    return df

def xls_to_csv(file):
    '''
    to read from this csv use:
    pd.read_csv(filepath, index_col=0)
    '''
    df = xls_to_pandas(file)
    filename = file #, _ = op.splitextl(file)
    csv_filename = filename + ".csv"
    df.to_csv(csv_filename)
    

def bed_to_series(f):
    pass

def find_clusters_in(a,b):
    '''
    find if clusters in a are also in b:
    
    An a cluster in is B if:
    - Same chromosoma
    - cluster initial value and end value are both in B
    '''
    
    
```
